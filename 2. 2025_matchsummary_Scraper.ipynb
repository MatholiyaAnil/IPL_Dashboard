```python
import requests
import json
import re
import pandas as pd
```

```python
test_url = 'https://ipl-stats-sports-mechanic.s3.ap-south-1.amazonaws.com/ipl/feeds/203-matchschedule.js'

response = requests.get(url=test_url)

start_index = response.text.find('(') + 1
end_index = response.text.rfind(')')

json_data = response.text[start_index:end_index]
data = json.loads(json_data)

match_schedule = data['Matchsummary']
```

```python
match_schedule
```

```python
df = pd.DataFrame(match_schedule )
```

```python
# Save DataFrame as CSV with the file name derived from the test URL
file_name = re.search(r'/(\d+-\w+)\.', test_url).group(1) 
csv_file_name = f"{file_name}-MatchSummary.csv"  # Append 'OverHistory' to the file name
df.to_csv(csv_file_name, index=False)

print(f"CSV file saved as {csv_file_name}")
```
